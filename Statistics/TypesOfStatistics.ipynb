## Covariance
Python implementation for covariance of two variables

```python
x = [2, 4, 6]
y = [1, 3, 5]

meanx = sum(x) / float(len(x))
meany = sum(y) / float(len(x))

xpart = [i -meanx for i in x]
ypart = [i - meany for i in y]

numerator = sum([xpart[i] * ypart[i] for i in range(len(xpart))])

denominator = len(x) - 1
covariance = numerator / denominator
```

## Correlation 
Python code implementation for correlation coefficient

```python
def covariance(x, y):
    meanx = sum(x) / float(len(x))
    meany = sum(y) / float(len(y))

    xpart = [i - meanx for i in x]
    ypart = [i - meany for i in y]

    numerator = sum(xpart[i] * ypart[i] for i in range(len(x)))
    return numerator / (len(x) - 1)


def standard_deviation(data):
    mean = sum(data) / float(len(data))
    squared_diff = [(i - mean) ** 2 for i in data]
    variance = sum(squared_diff) / (len(data) - 1)
    return variance ** 0.5


correlation = covariance(x, y) / (
    standard_deviation(x) * standard_deviation(y)
)
print(correlation)
```
