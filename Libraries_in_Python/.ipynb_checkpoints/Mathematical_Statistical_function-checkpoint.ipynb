### Common Mathematical and Statistical function in NumPy

```python
import numpy as np
```

```python
arr = np.array([
    [4, 3, 2],
    [10, 1, 0],
    [5, 8, 24]
])
```

```python
arr
```

```python
# amin() - minimum value in the array
np.amin(arr)
```

```python
# axis=0 --> means vertical minimum value
np.amin(arr, axis=0)
```

```python
# axis=1 --> means Horizontal minimum value
np.amin(arr, axis=1)
```

```python
# max for maxixum value in array
np.max(arr)
```

```python
np.amax(arr, axis=0)  # vertical maximum value
```

```python
np.amax(arr, axis=1)  # horizontal maximum value
```

### Similarly we have mediam, mean, standard deviation and variance
Everything can be used with access parameters
- axis=0 --> means vertical minimum value
- axis=1 --> means horizontal minimum value

```python
arr
```

```python
np.median(arr)
```

```python
np.mean(arr)
```

```python
# Standard deviation
np.std(arr)
```

```python
# variance
np.var(arr)
```

```python
# percentile value
np.percentile(arr, 50)  # percentile value 0 to 100
```

### Trigonometric function

```python
degree = np.array([0, 30, 45, 60, 90])
```

```python
np.sin(degree*np.pi/180)
```

```python
np.cos(degree*np.pi/180)
```

```python
np.tan(degree*np.pi/180)
```
