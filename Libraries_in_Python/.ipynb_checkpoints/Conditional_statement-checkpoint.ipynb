### Conditional statement in numpy

```python
import numpy as np
```

```python
x = np.array([i for i in range(10)])
```

```python
x
```

```python
# .where method for condition
np.where(x%2 == 0, 'Even', 'Odd')
```

```python
# .selection method for condition
conditionList = [x<5, x>5]
choiceList = [x**2, x**3]

np.select(conditionList, choiceList, default=x)
```
