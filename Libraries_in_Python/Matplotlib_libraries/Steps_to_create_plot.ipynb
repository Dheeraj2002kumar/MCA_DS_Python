## Steps to Create Plot - Example

#### Step 01: Import the required libraries

```python
# import numpy for generating random numbers
import numpy as np

# import matplotlib library
import matplotlib.pyplot as plt
from matplotlib import style
%matplotlib inline
```

## Step 02: Define or import the required dataset

```python
# Generate random numbers (total 10)
randomNumber = np.random.rand(10)
```

```python
# view them
print(randomNumber)
```

## Step 03: Set the plot parameters

```python
# Select the style of the plot
style.use('ggplot')

# plot the random number
plt.plot(randomNumber, 'g', label='line one', linewidth=2)

# x axis is number of random number (index)
plt.xlabel('Range')

# y axis is actual random number
plt.ylabel('Number')

# Title of the plot
plt.title('First Plot')

# Step 04: Display the created plot
plt.legend()
plt.show()
```
