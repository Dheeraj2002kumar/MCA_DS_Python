## Using Matplotlib to plot Graphs

```python
import pandas as pd
import matplotlib.pyplot as plt
```

```python
df = pd.read_csv('HousePrices.csv', parse_dates=['date'], index_col=['date'])
df.sort_index(inplace=True)
```

```python
df.head()
```

```python
plt.plot(df['price'])
```

```python
# plot boxplot
plt.boxplot(df['price'])
```

```python
# histogram
plt.hist(df['sqft_living'])
```

```python
# scatter plot
plt.xlabel('price')
plt.ylabel('sqft_living')
plt.title('Scatter Plot')
plt.scatter(x=df['price'], y = df['sqft_living'])
```

```python
# bar plot
d = {'a':10, 'b': 20, 'c':13}
plt.bar(x = d.keys(), height= d.values())
```

```python
# pie plot
plt.pie(x = d.values(), labels= d.keys())
```

```python
plt.plot(df['price'])
```

```python
plt.figure(figsize = (15, 10), dpi = 100)
plt.xlabel('Time')
plt.ylabel('Price')
plt.title('Line Plot of Prices')
plt.plot(df['price'])
```

```python
# save the line plot
plt.figure(figsize = (15, 10), dpi = 100)
plt.xlabel('Time')
plt.ylabel('Price')
plt.title('Line Plot of Prices')
plt.plot(df['price'])
plt.savefig('Prices.png')  # save the line plot
```
