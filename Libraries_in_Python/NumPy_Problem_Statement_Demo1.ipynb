## Problem Statement

Create a one-dimension array with alias as np and perform the following methods:

1. np.zeros()
2. np.ones()
3. np.empty()
4. np.arange()
5. np.reshape()
6. np.linspace()

```python
# import NumPy library
import numpy as np
```

```python
# first NumPy array
first_numpy_array = np.array([1, 2, 3, 4])

# Regular array: Mention array elements within brackets
```

```python
print(first_numpy_array)
```

```python
# array with zeros
# Zeros array: Mention shape of array within parentheses
array_with_zeros = np.zeros((3, 3))
array_with_zeros
```

```python
# Array with ones
array_with_ones = np.ones((3, 3))
array_with_ones
```

```python
# array with empty
array_with_empty = np.empty((2, 3))
array_with_empty
```

```python
# array with arange method
np_arange = np.arange(12)
```

```python
# print array
print(np_arange)
```

```python
# reshape method to change / create array
np_arange.reshape(3, 4)
```

```python
# linspace for linearly (equal) spaced data elements
np_linspce = np.linspace(1, 6, 5)

# First element = 1
# Last element = 6
# Total number of equidistant elements = 5
```

```python
# print linespaced array
print(np_linspce)
```

```python
# One dimensional array
oneD_array = np.arange(15)
```

```python
# print 1D array
print(oneD_array)
```

```python
# Two dimensional array
TwoD_array = oneD_array.reshape(3, 5)
```

```python
# print 2D array
print(TwoD_array)
```

```python
# Three dimensional array
ThreeD_array = np.arange(27).reshape(3, 3, 3)
```

```python
# Print 3D array
print(ThreeD_array)
```
