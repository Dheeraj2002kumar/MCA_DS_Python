### Arithmetic Operation with numpy

```python
import numpy as np
```

```python
A = np.array([
    [1, 2, 1], 
    [2, 2, 3]
])
B = np.array([3, 4, 5])
```

```python
A
```

```python
B
```

```python
# Add two Array
np.add(A, B)
```

```python
# Subtract
np.subtract(B, A)
```

```python
# Multipy
np.multiply(A, B)
```

```python
# Divide
np.divide(A, B)
```

```python
# Power of array element
np.power(A, 2)
```

```python
np.power(A, B)
```
