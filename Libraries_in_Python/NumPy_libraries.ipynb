## Shap Manipulation - Example
You can use certain function to manipulate the shape of an array to do the following:

```python
new_cyclist_trials = np.array([
    [10, 15, 17, 26, 13, 19],
    [12, 11, 21, 24, 14, 23]
])
```

```python
new_cyclist_trials.ravel()
```

```python
new_cyclist_trials.reshape(3, 4)
```

```python
new_cyclist_trials.resize(2, 6)
```

```python
new_cyclist_trials
```

```python
np.hsplit(new_cyclist_trials, 2)
```

```python
new_cyclist_1 = np.array([10, 15, 17, 26, 13, 19])
```

```python
new_cyclist_2 = np.array([12, 11, 21, 24, 14, 23])
```

```python
np.hstack((new_cyclist_1, new_cyclist_2))
```
