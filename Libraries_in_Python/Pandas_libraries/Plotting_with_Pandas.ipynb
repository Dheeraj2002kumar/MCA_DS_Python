### Ploting with pandas 

```python
import pandas as pd
```

```python
df = pd.read_csv('HousePrices.csv')
```

```python
df.head()
```

```python
df.columns
```

```python
df.plot()
```

```python
df['price'].plot()
```

```python
df['price'].plot(legend=True)
```

```python
# plot scatter plot
df.plot(x='price', y='bedrooms', kind='scatter', legend=True)
```

```python
# add figure size(x, y)
df.plot(title='Sample Plot', figsize=(15, 10))
```

```python
# box plot to pass value [kind='box']
df['price'].plot(kind='box', title='Box Plot')
```

```python
# Histogram plot to pass value [kind='hist']
df['price'].plot(kind='hist', title='Histogram Plot')
```

```python
# Line Plot  Price trend over time 
df.plot(x='date', y='price', kind='line', title='House Price Trend')
```

```python
# Scatter Plot  Price vs Living Area
df.plot(x='sqft_living', y='price', kind='scatter',
        title='Price vs Living Area')
```

```python
# Bar Plot  Average price by number of bedrooms
df.groupby('bedrooms')['price'].mean().plot(
    kind='bar', title='Average Price by Bedrooms'
)
```

```python
# Box Plot  Price distribution by bedrooms
df.boxplot(column='price', by='bedrooms')
```

```python
# Histogram  Price distribution
df['price'].plot(kind='hist', bins=30, title='Price Distribution')
```

```python
# KDE (Density) Plot  Smooth price distribution
df['price'].plot(kind='kde', title='Price Density')
```

```python
# Area Plot  Total sales over time
df.groupby('date')['price'].sum().plot(
    kind='area', title='Total Sales Over Time'
)
```

```python
# Pie Chart  Waterfront vs Non-waterfront 
df['waterfront'].value_counts().plot(
    kind='pie', autopct='%1.1f%%', title='Waterfront Houses'
)
```

```python
# Heatmap  Correlation between numeric features 

import seaborn as sns
import matplotlib.pyplot as plt

numeric_df = df.select_dtypes(include='number')

plt.figure(figsize=(10,6))
sns.heatmap(numeric_df.corr(), cmap='coolwarm')
plt.title('Feature Correlation Heatmap')
plt.show()
```

```python
# Pair Plot  Relationships between key features

sns.pairplot(df[['price', 'sqft_living', 'bedrooms', 'bathrooms']])
```
