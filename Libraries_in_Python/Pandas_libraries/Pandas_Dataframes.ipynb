<p><b>Note:</b> If the file size is very large, we can also read the file and chunk sizes as following </p>

```python
df = pd.read_csv('PandasExample.csv', chunksize=2) 
# return an iterable
```

```python
for chunk in df:
    print(chunk)
```

##### Comparison: with vs without chunksize

| Without `chunksize`           | With `chunksize=2`    |
| ----------------------------- | --------------------- |
| Entire CSV loaded into memory | Data loaded in parts  |
| `df` is a DataFrame           | `df` is an iterator   |
| Faster for small files        | Safer for large files |
