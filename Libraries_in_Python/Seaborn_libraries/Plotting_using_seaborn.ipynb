## Plotting Using Seaborn

```python
import seaborn as sns
```

```python
sns.set_theme()
fmri = sns.load_dataset("fmri")
sns.relplot(data = fmri, kind = "line", x = "timepoint", y = "signal",
           col = "region", hue = "event", style = "event",)
```
